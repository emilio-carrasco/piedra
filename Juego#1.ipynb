```python
def pide_num_jugadas():
    
    num = int(input((f"""¿A qué quieres jugar?\n\n3️⃣ Piedra, papel o tijera\n5️⃣ Piedra papel, tijera, lagarto o Spoke\n❓ Otro impar mayor o igual que 3:""")))
    while num % 2 == 0 or num < 3:
        num = int(input((f"""¿A qué quieres jugar?\n\n3️⃣ Piedra, papel o tijera\n5️⃣ Piedra papel, tijera, lagarto o Spoke\n❓ Otro impar mayor o igual que 3:""")))

    return num

def pide_nombre_jugadas(num):
#esto es mejorable ya que puedo asegurarme que todas las palabras son distintas
# tb podríamos pedir un emoticonos
    lista_jugadas  =[str(input(f"Dime el nombre de la jugada #{i}:")) for i in range(1, num + 1)] 
    return lista_jugadas

def crea_ranking(lista):
    from math import floor
    l = len(lista)
    a_quien_gana = dict() 
    for i, elemento in enumerate(lista):
        perdedores = set()
        for j in range(0, floor(l/2)):     
            perdedores.add(lista[(i + (1 + j * 2)) %l])
        a_quien_gana[elemento] = perdedores

    return a_quien_gana

def pide_jugada(lista):
    l = len(lista)
    print("")
    for i, jugada in enumerate(lista):
        print(f"#{i + 1} {jugada}")
        
    num = int(input(f"TU JUGADA:"))
    while num > l or num < 0:
        num = int(input(f"❌ TU JUGADA:"))
        
    return lista[num - 1]


def aleatorio(lista): 
    import random
    
    return random.choice(lista)



def imprime_marcador(marcador):
    print(f"\n➖➖ MARCADOR RONDA {marcador['ronda']} ➖➖")
    print(f"🚻HUMAN@:{marcador['victorias_humano']}       💻SKYNET:{marcador['victorias_skynet']}\n\n")

    
def imprime_resultado(marcador):
    
    from datetime import date
  
    if marcador['victorias_humano'] == marcador['victorias_skynet']:
        print(f"\n¡¡¡Hemos llegado a la última ronda y hay un empate!!!\n")                                               
    elif marcador['victorias_humano'] > marcador['victorias_skynet']:
        print(f"😤Gana el humano ¿no serás un replicante?🤖")
    else:
        print(f"{date.today()}:🦾 Skynet toma conciencia de sí mismo y te gana :P")
   

    
def imprime_jugada(jugada_humano,jugada_skynet):
    
    
    print(f"\n➖➖➖➖ JUGADA ➖➖➖➖")
    print(f"""
 🚻HUMAN@         💻SKYNET
{jugada_humano}  🆚   {jugada_skynet}\n""")    
    return

def iniciar_marcador(v,r_max):

    marcador={'victorias_humano':int(0), 'victorias_skynet':int(0), 'victorias_para_ganar':v, 'ronda':int(1), 'max_rondas':r_max+1 }
    return marcador


def pide_lista_jugadas(n):   

    if n == 5:
        print(f"\n ⚡ ⚡ ⚡ 🌑 PIEDRA 🧻 PAPEL ✂ TIJERA 🐲 LAGARTO 👽 SPOKE ⚡ ⚡ ⚡\n\n")
        l = ['✂ Tijeras','🧻 Papel','🌑 Piedra','🐲 Lagarto','👽 Spoke']
    elif n == 3:
        print(f"\n ⚡ ⚡ ⚡ 🌑 PIEDRA 🧻 PAPEL ✂TIJERA ⚡ ⚡ ⚡\n\n")
        l = ['✂ Tijeras','🧻 Papel','🌑 Piedra']
    else:
        print(f" 😕 No conozco el nombre de este juego, ¿cómo son las jugadas? \n\n")
        l = pide_nombre_jugadas(n)
    
    return l

def seguir(marcador):
    v = marcador['victorias_para_ganar']
    sigue  =marcador['victorias_humano'] < v and marcador['victorias_skynet'] < v and marcador['ronda'] < marcador['max_rondas']
    return sigue                        

    
def gana(j1, j2, marcador):
    return(j2 in marcador[j1])
    
def incrementa(marcador, string):
    marcador[string] += 1
    return marcador

def imprime_uno_a_uno(cadena,t):
    import time
    for letra in cadena:
        print(letra, end='')
        time.sleep(t)
    print("")
def imprime_num_ronda(marcador):
    
    cadena=(f"➖➖➖➖➖➖➖➖➖➖➖➖➖➖➖➖ RONDA #{marcador['ronda']} ➖➖➖➖➖➖➖➖➖➖➖➖➖➖➖➖")
    imprime_uno_a_uno(cadena,t=0.05)
    
def imprime_gana_humano():
    print(f"💥💥💥Gana HUMAN@ 🚻💥💥💥\n")   
    
def imprime_gana_skynet():
    print(f"💥💥💥Gana SKYNET 💻💥💥💥\n")  
    
def imprime_empate():
    print(f"⚖⚖⚖ EMPATE ⚖⚖⚖ \n")
    print(f"⚠ No corre ronda ;)")

    
def juego():    
    import time

    # Inicio del juego

    num_jugadas = pide_num_jugadas()                #hemos creado el mismo algoritmo para los 'n' juegos similares: https://www.microsiervos.com/archivo/humor/piedra-papel-tijera-pistola-rayo-dragon-agua-aire-esponja-lobo-arbol-humano-serpiente-fuego.html
    lista_jugadas = pide_lista_jugadas(num_jugadas) #en caso de que sea 3 y 5 sabemos cómo llamarlos pero para otros valores habría que pedir nombre de cada jugada

    # Creamos el diccionario con a quienes gana cada jugada
    ranking = crea_ranking(lista_jugadas)           #creamos un jugada:{perdedores} donde la clave es una de las jugadas y el valor es el set de todos a los que esa jugada gana

    #Iniciamos las variables de juego
    victorias_para_ganar = 3                        #tres victorias de un jugador termina el juego
    max_rondas = 10                                 #máximo número de jugadas ¡¡¡ojo: hemos implementado que los empates no cuentan ronda así que esto no tiene sentido
    tiempo_printing = 0.5
    marcador = iniciar_marcador(victorias_para_ganar,max_rondas)


    #bucle principal
    while seguir(marcador):
        imprime_num_ronda(marcador)               #imprimir num de ronda
        jugada_humano = pide_jugada(lista_jugadas)  #lo primero es pedirle la jugada al humano
        jugada_skynet = aleatorio(lista_jugadas)    #después sacamos el aleatorio de skynet
        
        imprime_jugada(jugada_humano, jugada_skynet) #sacamos por pantalla ambas jugadas
        time.sleep(3*tiempo_printing)
        
        if jugada_humano == jugada_skynet:             #vemos si es empate para no correr ronda
             imprime_empate()
        else:
            incrementa(marcador, 'ronda')             #incrementamos el num de ronda si no ha sido empate

            if gana(jugada_humano, jugada_skynet, ranking): #gana humano
                incrementa(marcador, 'victorias_humano')   #incrementamos su marcador
                imprime_gana_humano()                #imprimimos que gana humano
            else:
                incrementa(marcador, 'victorias_skynet')   #si gana skynet
                imprime_gana_skynet()                #imprimimos que gana skynet
        time.sleep(2*tiempo_printing)
        imprime_marcador(marcador)                        #imprimimos el estado del marcador
        time.sleep(2*tiempo_printing)

    imprime_resultado(marcador)                       #hemos terminado, ¿quién gana?



juego()
```
